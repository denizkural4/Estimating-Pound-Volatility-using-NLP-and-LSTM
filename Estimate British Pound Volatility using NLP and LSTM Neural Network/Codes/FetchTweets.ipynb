## Twitter API for Scrapping Tweets from Twitter

```python
import tweepy
#enter necessary keys
consumer_key = "YOUR CONSUMER KEY"
consumer_secret = "YOUR CONSUMER SECRET"
access_secret = "YOUR ACCESS SECRET"
access_token_secret = "YOUR ACCESS TOKEN SECRET"

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_secret, access_token_secret)
api = tweepy.API(auth)
```

### Building Training set for creating our corpus

#### Corpus is containing tweets about Brexit between Jan 2016 and September 2019. 

```python
def buildTrainingSet(corpusFile, tweetDataFile):
    import csv
    import time

    counter = 0
    corpus = []
    # 15 minutes time interval due API limits within 900

    with open(corpusFile, 'r') as csvfile:
        lineReader = csv.reader(csvfile, delimiter='~', quotechar="\"")
        for row in lineReader:
            corpus.append({"tweet_id":row[0], "user_id":row[1], "t_sentiment":row[2],"t_stance":row[3]}) 

    sleepTime = 2
    trainingDataSet = []

    for tweet in corpus:
        try:
            tweetFetched = api.get_status(tweet["tweet_id"])
            # fetching the user 
            user = api.get_user(tweet["user_id"]) 
            # fetching the followers_count 
            followers_count = user.followers_count
            print("Tweet fetched" + tweetFetched.text)
            tweet["text"] = tweetFetched.text
            tweet["date"]=tweetFetched.created_at.date()
            tweet["followers"]=followers_count
            trainingDataSet.append(tweet)
            time.sleep(sleepTime)

        except:
            print("Inside the exception")
            continue
    print("Nu. of Tweets fetched:")
    #Saving the fetched tweeets into csv
    with open(tweetDataFile, 'w') as csvfile:
        linewriter = csv.writer(csvfile, delimiter=',', quotechar="\"")
        i=1
        for tweet in trainingDataSet:
            try:
                linewriter.writerow([tweet["date"],tweet["tweet_id"], tweet["text"], tweet["t_sentiment"],tweet["followers"]])
                print("%s tweets fetched " % i)
                i+=1
            except Exception as e:
                print(e)
    return trainingDataSet
```
